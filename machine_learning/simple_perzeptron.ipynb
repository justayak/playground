```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

epsilon = 0.1

Training = np.array([
    [0,0,1],
    [2,2,-1],
    [-2+epsilon, 4 + epsilon, -1],
    [-2-epsilon, 4 - epsilon, 1],
    [4, -1-epsilon, -1],
    [4-epsilon, -1, 1],
    [1,2,1],
    [3,1,-1]
])

fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(111)
ax.plot([0, 0], [-3, 6], color='black', alpha=0.5)
ax.plot([-3, 5], [0,0], color='black', alpha=0.5)

for t in Training:
    if t[2] > 0:
        ax.scatter(t[0], t[1], color='green')
    else:
        ax.scatter(t[0], t[1], color='red')

plt.show()
```
