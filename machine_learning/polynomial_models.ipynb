```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from random import random

data = np.array([(x, x + 2*(0.5 - random())) for x in range(1, 10, 2)])

data = np.array([
    (0, 0.1),(1, 1.2), 
    (2, 1.98),
    (3.01, 3), (3, 3.5), (3.3, 3), (2.5, 3), 
    (7.5, 8.5), (8, 7.55), (8.2, 7.3),
    (9, 9.11) ])

lin = np.polyfit(data[:,0], data[:,1], 1)
pol = np.polyfit(data[:,0], data[:,1], 10)

def fun(Z, x):
    result = 0
    for i, z in enumerate(reversed(Z)):
        result += z * x**i
    return result

linear = np.poly1d(lin)
poly = np.poly1d(pol)

linear = np.array([(x, linear(x)) for x in range(10)])
poly = np.array([(x, poly(x)) for x in range(10)])



fig = plt.figure(figsize=(16,3))
ax = fig.add_subplot(121)
ax.scatter(data[:,0], data[:,1])
ax.plot(linear[:,0], linear[:,1])
ax.set_yscale('log')
ax.set_y

ay = fig.add_subplot(122)
ay.scatter(data[:,0], data[:,1])
ay.plot(poly[:,0], poly[:,1])
#ay.set_yscale('log')
```

```python
fig = plt.figure(figsize=(4,1))
ax = fig.add_subplot(111)
ax.scatter(data[:,0], data[:,1], color='gray')
line_up, = ax.plot(linear[:,0], linear[:,1], label="P(x)^1", ls='--', color='black')
line_down, = ax.plot(poly[:,0], poly[:,1], label='P(X)^10', color='lightgray')
#ax.set_yscale('log')

plt.legend(handles=[line_up, line_down])
#plt.show()

plt.savefig('polyfit.eps', format='eps', dpi=1000)
```
