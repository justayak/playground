## Filters

**Gamma correction:**: $g(x)=\big[f(x)\big]^{1/\gamma}$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2, sys
import numpy as np
sys.path.insert(0, '..')
import helper

I = cv2.cvtColor(cv2.imread('../smila.jpg'), cv2.COLOR_BGR2GRAY)

def gamma_correct(I, gamma):
    I_1 = np.power(I, 1/gamma)
    I_1 = I_1 / np.max(I_1)
    return (I_1 * 255).astype('uint8')

helper.plot_mats([ 
    gamma_correct(I, 1.9), I, gamma_correct(I, 0.2)], cols=3)
```
